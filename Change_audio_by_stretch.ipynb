```python
import librosa
import os
import librosa.display
import matplotlib.pyplot as plt
import IPython.display
from scipy.io.wavfile import write
```

```python
sample,sr = librosa.load('C:/Users/ADmin/Desktop/low_mid.wav')
```

```python
plt.plot(sample)
```

```python
stretch = librosa.effects.time_stretch(sample,2.0)
shrink = librosa.effects.time_stretch(sample,0.5)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1),plt.plot(sample)
plt.subplot(1,3,2),plt.plot(stretch)
plt.subplot(1,3,3),plt.plot(shrink)
```
