```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
wav_k,_ = librosa.load("C:/Users/ADmin/Desktop/오목눈이/kick.wav")
wav_c,_ = librosa.load("C:/Users/ADmin/Desktop/오목눈이/crash.wav")
wav_k_c,_ = librosa.load("C:/Users/ADmin/Desktop/오목눈이/kick+crash.wav")
```

```python
stft_k = np.abs(librosa.stft(wav_k))
stft_c = np.abs(librosa.stft(wav_c))
stft_k_c = np.abs(librosa.stft(wav_k_c))
stft_sum = np.abs(librosa.stft(wav_k+wav_c))
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
librosa.display.specshow(stft_k, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('kick')
plt.subplot(4, 1, 2)
librosa.display.specshow(stft_c, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('crash')
plt.subplot(4, 1, 3)
librosa.display.specshow(stft_k_c, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('kick_crash')
plt.subplot(4, 1, 4)
librosa.display.specshow(stft_sum, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('kick+crash')
```

```python
data = [stft_k, stft_c, stft_k_c, stft_sum]
val = np.amax(data)

plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
librosa.display.specshow(librosa.amplitude_to_db(stft_k, ref=val), y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('kick')
plt.subplot(4, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(stft_c, ref=val), y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('crash')
plt.subplot(4, 1, 3)
librosa.display.specshow(librosa.amplitude_to_db(stft_k_c, ref=val), y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('kick_crash')
plt.subplot(4, 1, 4)
librosa.display.specshow(librosa.amplitude_to_db(stft_sum, ref=val), y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('kick+crash')
```

```python
k_250 = stft_k[250,:]
k_500 = stft_k[500,:]
k_750 = stft_k[750,:]
k_900 = stft_k[900,:]

c_250 = stft_c[250,:]
c_500 = stft_c[500,:]
c_750 = stft_c[750,:]
c_900 = stft_c[900,:]

k_c_250 = stft_k_c[250,:]
k_c_500 = stft_k_c[500,:]
k_c_750 = stft_k_c[750,:]
k_c_900 = stft_k_c[900,:]

sum_250 = stft_sum[250,:]
sum_500 = stft_sum[500,:]
sum_750 = stft_sum[750,:]
sum_900 = stft_sum[900,:]

plt.figure(4,[20,12])
plt.subplot(221)
plt.title('kick')
plt.plot(k_250, color='red', label='250')
plt.plot(k_500, color='green', label='500')
plt.plot(k_750, color='blue', label='750')
plt.plot(k_900, color='orange', label='1000')
plt.legend()
plt.grid()

plt.subplot(222)
plt.title('crash')
plt.plot(c_250, color='red', label='250')
plt.plot(c_500, color='green', label='500')
plt.plot(c_750, color='blue', label='750')
plt.plot(c_900, color='orange', label='1000')
plt.legend()
plt.grid()

plt.subplot(223)
plt.title('kick_crash')
plt.plot(k_c_250, color='red', label='250')
plt.plot(k_c_500, color='green', label='500')
plt.plot(k_c_750, color='blue', label='750')
plt.plot(k_c_900, color='orange', label='1000')
plt.legend()
plt.grid()

plt.subplot(224)
plt.title('kick+crash')
plt.plot(sum_250, color='red', label='250')
plt.plot(sum_500, color='green', label='500')
plt.plot(sum_750, color='blue', label='750')
plt.plot(sum_900, color='orange', label='1000')
plt.legend()
plt.grid()

plt.show()
```
